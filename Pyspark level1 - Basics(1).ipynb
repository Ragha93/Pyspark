```
####3. Filtering columns with single and multiple conditions, filter and collect to convert it to dictionary
```

```
new_df.filter('Open > 300').show(5)
```

```
new_df.filter((new_df['Open']>=309) & (new_df['Close']>=307)).show(5)
```

```
new_df.filter((new_df['Open']>300) & (new_df['High']>300)).select(['High','Close']).show(5)
```

```
x = new_df.filter(new_df['Open']>=200).select(['Date','Volume','Open']).collect()[1].asDict()
```

```
df = SS1.read.table('sales_info')
#Read data directly from a table

df1 = SS1.sql('select * from sales_info')
#read data using SQL query
```

```
df1 = df1.withColumn('Sales',df1['Sales'].cast(IntegerType()))
```

```
df1.printSchema()
```

```
df1.groupBy('Company').mean().show()
```

```
df1.agg({'Sales':'sum'}).show()
```

```
from pyspark.sql.functions import countDistinct,avg,stddev,format_number

df1.select(countDistinct('sales')).show()
```

```
df1.select(avg('Sales').alias('Average_sales')).show()
```

```
df1.select(format_number(avg('Sales'),2).alias("Average_Sales")).show()
```

```
df1.orderBy('Sales').show()
```

```
df1.orderBy('Sales',ascending=False).show()
```

```
df1.orderBy(df1['Sales'].desc()).show()
```

```
ss2 = spark.builder.appName("New").getOrCreate()
df3 = ss2.read.csv("/FileStore/tables/ContainsNull.csv",inferSchema=True,header=True)
```

```
df3.printSchema()
```

```
df3.na.drop().show()
```

```
df3.na.drop(thresh=2).show()
```

```
df3.na.fill('Noname').show()
```

```
from pyspark.sql.functions import mean
mean_val = df3.select(mean('Sales')).collect()[0][0]
```

```
df3.fillna(mean_val,subset=['Sales']).show()
```

```
df3.na.fill(df3.select(mean('Sales')).collect()[0][0],['Sales']).show()
```
